# Data Analysis ans Visualization

```python
canceled_percent = df['is_canceled'].value_counts(normalize = True)
print('canceled_percent')

plt.figure(figsize=(5,4))
plt.title('Reservation Status Count')
plt.bar(['Not Canceled', 'Canceled'], df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.7)
plt.show()

```

```python
plt.figure(figsize=(8,4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(loc='upper right', bbox_to_anchor=(1,1))
plt.title('Reservation Satus in different hotels', size = 20)
plt.xlabel('Hotel')
plt.ylabel('Number Of Reservation')
plt.legend(['Not_Canceled','Canceled'])
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort_Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City_Hotel')
plt.legend(fontsize = 15)
plt.show()


```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
as1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'bright')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Satus per Month', size = 20)
plt.xlabel('Month')
plt.ylabel('Number Of Reservation')
plt.legend(['Not_Canceled','Canceled'])
plt.show()
```

```python
plt.figure(figsize = (15,8))
plt.title('ADR per month', fontsize = 30)
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index(), palette = 'Blues', hue = 'month')
plt.show()



```

```python
import matplotlib.pyplot as plt

canceled_data = df[df['is_canceled'] == 1]
top_10_countries = canceled_data['country'].value_counts()[:10]

plt.figure(figsize=(8, 8))
plt.title('Top 10 Countries with Reservations Canceled')
plt.pie(top_10_countries, autopct='%.2f', labels=top_10_countries.index)

plt.show()

```
